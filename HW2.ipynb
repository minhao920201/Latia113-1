承上學生成績資訊，計算出各科的平均數、中位數、標準差

```python
import pandas as pd
import numpy as np

#承上資料
student_ids = ['A{:03d}'.format(i) for i in range(1, 17)]

np.random.seed(42)
chinese_scores = np.random.randint(60, 101, size=16)
english_scores = np.random.randint(60, 101, size=16)
math_scores = np.random.randint(60, 101, size=16)


data = {'Student ID': student_ids,
     'Chinese': chinese_scores,
     'English': english_scores,
     'Math': math_scores
}
df = pd.DataFrame(data)
display(df)
#-------------------------------------------------------------------------------

# 計算各科的平均值、中位數和標準差
numeric_columns = ['Chinese', 'English', 'Math']
average_scores = df[numeric_columns].mean()
median_scores = df[numeric_columns].median()
std_scores = df[numeric_columns].std()

# 列出各科的平均值、中位數和標準差
print("各科平均值：")
print(average_scores)
print("\n各科中位數：")
print(median_scores)
print("\n各科標準差：")
print(std_scores)
```

```python
def a(x, y):
  result = x + y * 2
  return result
b = a(1, 2)
c = a(2, 3)
print(b, c)
```

```python
def a(x):
  result = 0
  while result < 10:
    result = result + x
    if result == 5:
      return result
  return result

b = a(1)
c = a(2)
print(b, c)
```

```python
def tf(value):
  return 2 * value - 1
test_values = [0.0, 0.0011, 0.25, 0.5, 0.54, 0.75, 1.0]
scaled_values = [tf(value) for value in test_values]
print('Original values:', test_values)
print('Scaled values:', scaled_values)
```

```python
def a(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n + a(n-1)

print(a(3))
```
