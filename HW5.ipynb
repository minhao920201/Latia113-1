Homework 2 (畫出 A, B, C, D 四張圓餅圖)

```python
import numpy as np
import pandas as pd
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))
#display(df)

import matplotlib.pyplot as plt
# 繪製每一個欄位的圓餅圖
for column in df.columns:
  data = df[column].abs()  # 使用絕對值來處理負數
  plt.figure(figsize=(6, 6))
  plt.pie(data, labels=df.index.strftime("%Y-%m-%d"), autopct='%1.1f%%', startangle=140)
  plt.title(f"Pie Chart for Column {column}")
  plt.show()
```

Homework 2 (畫出 A, B, C, D 四張3D柱狀長條圖)

```python
import numpy as np
import pandas as pd
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))
#display(df)

import matplotlib.pyplot as plt
# 設置 3D 繪圖
fig = plt.figure(figsize=(12, 10))

# 繪製每個欄位的 3D 柱狀圖
for idx, column in enumerate(df.columns, 1):
  ax = fig.add_subplot(2, 2, idx, projection='3d')  # 2x2 網格佈局

  # 資料的 x 軸為時間，y 軸為索引，z 軸為數值
  x = np.arange(len(df))  # x 軸為日期的索引
  y = np.zeros_like(x)  # y 軸為0，柱狀圖可以根據需要調整
  z = np.zeros_like(x)  # z 軸為0，表示從基線開始繪製柱狀圖

  # 3D 柱狀圖
  ax.bar3d(x, y, z, 1, 1, df[column], shade=True)

  # 設定標題和軸標籤
  ax.set_title(f"3D Bar Chart - {column}")
  ax.set_xlabel("Index")
  ax.set_ylabel("Zeroed Y")
  ax.set_zlabel(f"{column} values")

plt.tight_layout()
plt.show()
```

畫出相關係數圖

```python
import numpy as np
import pandas as pd
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))
#display(df)


import seaborn as sns
import matplotlib.pyplot as plt

# 計算相關係數矩陣
corr = df.corr()

# 使用seaborn畫出熱圖
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

產生ANOVA，從p-value可發現數據之間有顯著差異

```python
import numpy as np
import pandas as pd
from scipy import stats
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))
#display(df)


# 使用 scipy.stats.f_oneway 進行單因子變異數分析
f_statistic, p_value = stats.f_oneway(df['A'], df['B'], df['C'], df['D'])

# 顯示結果
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# 解釋結果
if p_value < 0.05:
  print("拒絕虛無假設，數據之間有顯著差異")
else:
  print("無法拒絕虛無假設，數據之間沒有顯著差異")
```

```python
import numpy as np
from scipy import stats
# 三個班學生的成績
group1 = [85, 86, 88, 75, 78, 94]
group2 = [92, 94, 89, 96, 93, 95, 76, 50]
group3 = [78, 81, 79, 84, 77, 80]
# 使用scipy.stats.f_oneway 進行單因子 ANOVA 檢定
f_statistic, p_value = stats.f_oneway(group1, group2, group3)
# 輸出結果
print("F 統計量:", f_statistic)
print("p-value:", p_value)
# 解釋結果
alpha = 0.05  # 顯著性水平
if p_value < alpha:
  print("拒絕原假設：不同組別之間的均值存在顯著差異。")
else:
  print("未能拒絕原假設：不同組別之間的均值沒有顯著差異。")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
# 創建數據集
data = {
 'Score': [85, 86, 88, 75, 78, 94, 92, 94, 89, 96, 93, 95, 78, 81, 79, 84, 77, 80],
 'Method': ['A', 'A', 'A', 'A', 'A', 'A',  # 教學方法 A
       'B', 'B', 'B', 'B', 'B', 'B',  # 教學方法 B
       'C', 'C', 'C', 'C', 'C', 'C']  # 教學方法 C
 }
# 將數據轉換為DataFrame
df = pd.DataFrame(data)
# 使用statsmodels 中的ols 方法進行單因子ANOVA 分析
model = ols('Score ~ C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
# 輸出ANOVA 表
print(anova_table)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Generate 100 rows of data
data = {
    'Score': np.random.randint(0, 101, 100),  # Random integers between 0 and 100
    'Method': np.random.choice(['A', 'B', 'C'], 100)  # Random selection of 'A', 'B', or 'C'
}

# Create a DataFrame
df = pd.DataFrame(data)
#display(df.head())

# Create a bar plot showing the average score for each method
plt.figure(figsize=(8, 6))
avg_scores = df.groupby('Method')['Score'].mean()
avg_scores.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Average Score by Method')
plt.xlabel('Method')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()

# 使用statsmodels 中的ols 方法進行單因子ANOVA 分析
model = ols('Score ~ C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
# 輸出ANOVA 表
print(anova_table)
```
